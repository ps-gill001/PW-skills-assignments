Q3. What is multiple inheritance? Write a python code to demonstrate multiple

Multiple inheritance is a feature in object-oriented programming where a class can inherit attributes and methods from more than one parent class. This allows the child class to combine and reuse the functionality of the parent classes.

Here's an example in Python to demonstrate multiple inheritance:

```python
class ParentClass1:
    def __init__(self, name):
        self.name = name

    def display_name(self):
        print(f"Name from ParentClass1: {self.name}")

class ParentClass2:
    def __init__(self, age):
        self.age = age

    def display_age(self):
        print(f"Age from ParentClass2: {self.age}")

class ChildClass(ParentClass1, ParentClass2):
    def __init__(self, name, age):
        ParentClass1.__init__(self, name)
        ParentClass2.__init__(self, age)

    def display_info(self):
        self.display_name()
        self.display_age()

# Example usage
child_object = ChildClass("John", 30)
child_object.display_info()
```

In this example, the `ChildClass` inherits from both `ParentClass1` and `ParentClass2`. The `ChildClass` constructor calls the constructors of the parent classes to initialize the `name` and `age` attributes.

The `display_info` method of the `ChildClass` calls the `display_name` method from `ParentClass1` and the `display_age` method from `ParentClass2`, demonstrating the use of multiple inheritance.

When you run the code, the output will be:

```
Name from ParentClass1: John
Age from ParentClass2: 30
```

The key points to remember about multiple inheritance in Python are:

1. A child class can inherit from multiple parent classes.
2. The child class can access and use the methods and attributes of all its parent classes.
3. If there is a name conflict between the parent classes, the method resolution order (MRO) determines which method is called.
4. It's important to be careful when using multiple inheritance to avoid complexity and potential issues, such as the diamond problem.

Getter and setter methods in Python are used to control the access and modification of an object's attributes. They provide a way to encapsulate the internal representation of an object and add additional logic or validation when accessing or modifying its attributes.

Here's an example of a class with a getter and a setter method:

```python
class Person:
    def __init__(self, name, age):
        self._name = name
        self._age = age

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, new_name):
        if isinstance(new_name, str) and len(new_name.strip()) > 0:
            self._name = new_name
        else:
            print("Invalid name. Name must be a non-empty string.")

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, new_age):
        if isinstance(new_age, int) and new_age > 0:
            self._age = new_age
        else:
            print("Invalid age. Age must be a positive integer.")

# Example usage
person = Person("John", 30)
print(person.name)  # Output: John
person.name = "Jane"
print(person.name)  # Output: Jane

person.age = 35
print(person.age)  # Output: 35
person.age = -10  # Output: Invalid age. Age must be a positive integer.
```

In this example, the `Person` class has two instance variables: `_name` and `_age`.The `@property` decorator is used to define the `name` and `age` getter methods, which return the corresponding instance variables.

The `@name.setter` and `@age.setter` decorators are used to define the setter methods for the `name` and `age` attributes, respectively. These setter methods allow you to set the values of the instance variables with additional validation logic.

#Q5.What is method overriding in python? Write a python code to demonstrate method overriding.

Method overriding in Python is a feature that allows a subclass to provide its own implementation of a method that is already defined in its superclass. When the subclass calls the method, it will execute the implementation defined in the subclass, rather than the implementation in the superclass.

Here's an example to demonstrate method overriding in Python:

```python
class Animal:
    def make_sound(self):
        print("The animal makes a sound")

class Dog(Animal):
    def make_sound(self):
        print("The dog barks")

class Cat(Animal):
    def make_sound(self):
        print("The cat meows")

# Example usage
animal = Animal()
animal.make_sound()  # Output: The animal makes a sound

dog = Dog()
dog.make_sound()  # Output: The dog barks

cat = Cat()
cat.make_sound()  # Output: The cat meows
```

In this example, we have a base class `Animal` with a `make_sound` method. The `Dog` and `Cat` classes inherit from `Animal` and both override the `make_sound` method with their own implementations.

When we create instances of the `Animal`, `Dog`, and `Cat` classes and call the `make_sound` method, the appropriate implementation is executed based on the object's type.

The key points about method overriding in Python are:

1. The subclass must have a method with the same name as the method in the superclass.
2. The method in the subclass must have the same number and type of parameters as the method in the superclass.
3. Method overriding allows the subclass to provide its own implementation of a method, which will be used instead of the implementation in the superclass.
4. Method overriding is a way to achieve polymorphism in object-oriented programming, where objects of different classes can be treated as objects of a common superclass.
