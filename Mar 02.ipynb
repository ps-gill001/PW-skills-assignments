```python
'''Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.'''

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

#Note: Also add title, xlabel, and ylabel to the plot.
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
'''Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:'''

import numpy as np

#For line 1:
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

#For line 2:
y2 = np.array([50, 20, 40, 20, 60, 70])

#For line 3:
y3 = np.array([10, 20, 30, 40, 50, 60])

#For line 4:
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title('Line 1')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.figure(figsize=(10,8))
plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title('Line 2')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.figure(figsize=(10,8))
plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('Line 3')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.figure(figsize=(10,8))
plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title('Line 4')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
#Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.bar(x = company, height = profit)
plt.title('Company Profits - Bar Plot')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')

plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.barh(company,profit,color = 'green')
plt.title('Company Profits - Horizontal Bar Plot')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')

plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating the box plot
plt.figure(figsize=(8, 6))

plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Two Distributions')
plt.xlabel('Data Sets')
plt.ylabel('Values')

# Display the plot
plt.show()
```
