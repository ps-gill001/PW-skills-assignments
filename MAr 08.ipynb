Q1. What is the Probability density function?
  
The probability density function (PDF) is a fundamental concept in probability theory and statistics. It is a function that describes the relative likelihood of a random variable taking on a specific value within a given interval.

The PDF is denoted as f(x) and has the following properties:

1. The PDF is always non-negative, meaning f(x) ≥ 0 for all values of x.
2. The total area under the PDF curve over the entire range of the random variable is equal to 1. This represents the total probability of the random variable taking on any possible value.
3. The probability of the random variable falling within a specific interval [a, b] is given by the integral of the PDF over that interval: P(a ≤ X ≤ b) = ∫(a to b) f(x) dx.

The PDF is used to characterize the distribution of a continuous random variable. It provides a complete description of the probability distribution of the random variable, as opposed to the cumulative distribution function (CDF), which gives the probability of the random variable being less than or equal to a specific value.

The shape of the PDF curve can take various forms, such as the normal (Gaussian) distribution, the exponential distribution, the Poisson distribution, and many others, depending on the characteristics of the random variable being studied.

Understanding and working with probability density functions is essential in various fields, including probability theory, statistics, physics, engineering, and finance, where the analysis of continuous random variables is crucial for modeling and decision-making processes.

Q2. What are the types of Probability distribution?

There are several types of probability distributions, each with its own characteristics and applications. Some of the most common types include:

1. **Discrete Probability Distributions**:
   - **Bernoulli Distribution**: Represents the probability of success or failure in a single trial, where success has a probability \( p \) and failure has a probability \( 1 - p \).
   - **Binomial Distribution**: Describes the number of successes in a fixed number of independent Bernoulli trials.
   - **Poisson Distribution**: Models the number of events occurring in a fixed interval of time or space, given a known average rate of occurrence.

2. **Continuous Probability Distributions**:
   - **Normal Distribution (Gaussian Distribution)**: One of the most widely used distributions, characterized by a symmetric bell-shaped curve. It is often used to model real-world phenomena due to the central limit theorem.
   - **Uniform Distribution**: All outcomes in a given range are equally likely. It has a constant probability density over the range of possible values.
   - **Exponential Distribution**: Models the time between events in a Poisson process, where events occur continuously and independently at a constant average rate.
   - **Gamma Distribution**: Generalizes the exponential distribution and is used to model waiting times, survival times, and other positive continuous variables.

3. **Multivariate Probability Distributions**:
   - **Multinomial Distribution**: Generalizes the binomial distribution to multiple categories. It describes the probability of observing each possible combination of outcomes in multiple independent trials.
   - **Multivariate Normal Distribution**: Generalizes the normal distribution to multiple dimensions. It is characterized by a mean vector and a covariance matrix and is widely used in multivariate statistics.

4. **Empirical Probability Distributions**:
   - **Empirical Distribution**: Based on observed data rather than a theoretical model. It assigns a probability to each observed value, making it useful for non-parametric statistical analysis.

These are just a few examples of probability distributions, and there are many others used in various fields of science, engineering, finance, and social sciences. The choice of distribution depends on the characteristics of the data and the specific problem being addressed.

```python
'''Q3. Write a Python function to calculate the probability density function of a normal distribution with
given mean and standard deviation at a given point.'''

import math

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) of a normal distribution
    at a given point x with mean and standard deviation.

    Parameters:
    x (float): The point at which to evaluate the PDF.
    mean (float): The mean of the normal distribution.
    std_dev (float): The standard deviation of the normal distribution.

    Returns:
    float: The value of the PDF at the given point x.
    """
    coefficient = 1 / (math.sqrt(2 * math.pi * std_dev**2))
    exponent = -((x - mean)**2) / (2 * std_dev**2)
    pdf_value = coefficient * math.exp(exponent)
    return pdf_value

# Example usage:
mean = 0
std_dev = 1
x = 1
pdf_at_x = normal_pdf(x, mean, std_dev)
print("PDF at x =", x, ":", pdf_at_x)

```

Q4. What are the properties of Binomial distribution? Give two examples of events where binomial
distribution can be applied.

The binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials, where each trial has only two possible outcomes: success or failure. The properties of the binomial distribution include:

1. **Fixed Number of Trials (n)**: The number of trials, denoted by \( n \), is fixed in advance.

2. **Independent Trials**: Each trial is independent of the others. The outcome of one trial does not affect the outcome of another.

3. **Two Possible Outcomes**: Each trial results in one of two possible outcomes: success or failure.

4. **Constant Probability of Success (p)**: The probability of success, denoted by \( p \), is constant for each trial.

5. **Discrete Outcome**: The binomial random variable represents a count of successes out of \( n \) trials, and therefore, it can only take on integer values.

The probability mass function (PMF) of the binomial distribution is given by:

\[ P(X = k) = \binom{n}{k} \cdot p^k \cdot (1 - p)^{n-k} \]

Where:
- \( P(X = k) \) is the probability of getting exactly \( k \) successes in \( n \) trials.
- \( \binom{n}{k} \) is the binomial coefficient, representing the number of ways to choose \( k \) successes out of \( n \) trials.
- \( p \) is the probability of success in each trial.
- \( 1 - p \) is the probability of failure in each trial.
- \( k \) ranges from 0 to \( n \).

Examples of events where the binomial distribution can be applied include:

1. **Coin Flipping**: Consider flipping a fair coin \( n \) times and counting the number of heads. Each flip is independent, with a fixed probability of success (getting heads) of \( p = 0.5 \).

2. **Drug Trials**: In a clinical drug trial, participants are either given the drug (success) or a placebo (failure). The outcome of each participant is independent, and the probability of success (the drug being effective) is constant for each participant. The binomial distribution can be used to model the number of successful outcomes (e.g., patients cured) out of a fixed number of participants.

```python
'''Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point'''

from scipy.stats import poisson

def poisson_cdf(mean, k):
    """
    Calculate the cumulative distribution function (CDF) of a Poisson distribution
    with a given mean at a given point.

    Parameters:
    mean (float): The mean (λ) of the Poisson distribution.
    k (int): The point at which to evaluate the CDF.

    Returns:
    float: The value of the CDF at the given point k.
    """
    return poisson.cdf(k, mean)

# Example usage:
mean = 3  # Mean (λ) of the Poisson distribution
k = 5  # Point at which to evaluate the CDF
cdf_value = poisson_cdf(mean, k)
print(f"CDF of Poisson distribution with mean {mean} at point {k}: {cdf_value}")
```

#Q9. How mean and variance are related in Binomial distribution and Poisson distribution?

In probability theory and statistics, the mean and variance of the Binomial and Poisson distributions are closely related, but they follow different rules based on their respective parameters. Here’s how they are related for each distribution:

### Binomial Distribution

For a Binomial distribution with parameters \( n \) (number of trials) and \( p \) (probability of success), the mean (\(\mu\)) and variance (\(\sigma^2\)) are given by:

- **Mean**: 
  \[
  \mu = np
  \]
  The mean is the product of the number of trials and the probability of success.

- **Variance**:
  \[
  \sigma^2 = np(1 - p)
  \]
  The variance is the product of the number of trials, the probability of success, and the probability of failure (\(1 - p\)).

### Poisson Distribution

For a Poisson distribution with parameter \(\lambda\) (the average rate of occurrence or the mean number of events in a given interval), the mean and variance are both equal to \(\lambda\):

- **Mean**: 
  \[
  \mu = \lambda
  \]
  The mean is the parameter \(\lambda\).

- **Variance**:
  \[
  \sigma^2 = \lambda
  \]
  The variance is also the parameter \(\lambda\).

### Summary of Relationships

1. **Binomial Distribution**:
   - Mean: \( \mu = np \)
   - Variance: \( \sigma^2 = np(1 - p) \)
   - The variance is dependent on both the probability of success \( p \) and the probability of failure \( 1 - p \).

2. **Poisson Distribution**:
   - Mean: \( \mu = \lambda \)
   - Variance: \( \sigma^2 = \lambda \)
   - The mean and variance are equal and are determined by the parameter \(\lambda\).

These relationships highlight that while the Binomial distribution’s mean and variance depend on two parameters (\(n\) and \(p\)), the Poisson distribution’s mean and variance are both governed by a single parameter (\(\lambda\)).


Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

In a normal distribution, the least frequent data points appear in the tails of the distribution, which are farthest from the mean. Here’s a detailed explanation:

### Characteristics of a Normal Distribution

1. **Symmetry**:
   - A normal distribution is symmetric about its mean. This means that the frequency of data points is highest at the mean and decreases symmetrically as you move away from the mean.

2. **Mean, Median, and Mode**:
   - In a normal distribution, the mean, median, and mode are all located at the center of the distribution.

3. **Standard Deviation**:
   - The spread of the data is determined by the standard deviation (\(\sigma\)). Most of the data (about 68%) lies within one standard deviation from the mean, about 95% lies within two standard deviations, and about 99.7% lies within three standard deviations.

### Least Frequent Data Points

- **Tails of the Distribution**:
  - The least frequent data points are found in the tails of the distribution, which are the regions farthest from the mean.
  - As you move further away from the mean (in either direction), the frequency of the data points decreases exponentially.

- **Mathematical Perspective**:
  - The probability density function (PDF) of a normal distribution is given by:
    \[
    f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \cdot e^{-\frac{(x-\mu)^2}{2\sigma^2}}
    \]
  - Here, \( \mu \) is the mean and \( \sigma \) is the standard deviation.
  - The exponential term \( e^{-\frac{(x-\mu)^2}{2\sigma^2}} \) shows that as \( |x - \mu| \) increases, \( f(x) \) (the frequency or probability density) decreases rapidly.

### Visual Representation

In a bell-shaped normal distribution curve:
- The center of the curve, where the peak is located, represents the mean (\(\mu\)).
- The height of the curve decreases as you move away from the mean towards the left or right.
- The tails of the curve, which extend infinitely but approach zero, represent the regions where the least frequent data points are located.

### Conclusion

In a normal distribution, the least frequent data points appear in the tails, farthest from the mean. This is because the probability density decreases as the distance from the mean increases, leading to fewer data points in these regions.
