Q1: The main difference between a t-test and a z-test is the underlying assumption about the population standard deviation.

T-test:
- Used when the population standard deviation is unknown.
- Assumes the sample is drawn from a normal distribution.
- Appropriate for small sample sizes (n &lt; 30).
Example scenario: Comparing the mean test scores of two groups of students when the population standard deviation is unknown.

Z-test:
- Used when the population standard deviation is known.
- Assumes the sample is drawn from a normal distribution.
- Appropriate for large sample sizes (n ≥ 30).
Example scenario: Determining if the mean weight of a product produced by a factory is different from the target weight, given the known population standard deviation.

Q2: The difference between one-tailed and two-tailed tests lies in the direction of the alternative hypothesis.

One-tailed test:
- The alternative hypothesis specifies a direction (greater than or less than) for the difference.
- The test is conducted in a single direction (left-tailed or right-tailed).
Example: Testing if a new drug reduces blood pressure (alternative hypothesis: mean decrease &gt; 0).

Two-tailed test:
- The alternative hypothesis does not specify a direction for the difference.
- The test is conducted in both directions (left-tailed and right-tailed).
Example: Testing if a new teaching method affects student test scores (alternative hypothesis: mean scores are different).

Q3: Type 1 and Type 2 errors in hypothesis testing:

Type 1 error (α):
- Rejecting the null hypothesis when it is true.
- Example: Concluding that a new drug is effective when it is not.

Type 2 error (β):
- Failing to reject the null hypothesis when it is false.
- Example: Concluding that a new drug is not effective when it is.

The significance level (α) is the probability of a Type 1 error, and the power of a test is 1 - β, which is the probability of detecting an effect if it exists.

Q4: Bayes' Theorem is a mathematical formula that relates conditional probabilities. It is used to update the probability of an event occurring based on new evidence.

The formula for Bayes' Theorem is:
P(A|B) = (P(B|A) * P(A)) / P(B)

Where:
- P(A|B) is the conditional probability of event A occurring given that event B has occurred.
- P(B|A) is the conditional probability of event B occurring given that event A has occurred.
- P(A) is the prior probability of event A occurring.
- P(B) is the prior probability of event B occurring.

Example: A medical test has a 90% chance of giving a positive result for a disease when the person has the disease (P(positive|disease) = 0.9), and a 5% chance of giving a positive result when the person does not have the disease (P(positive|no disease) = 0.05). If 1% of the population has the disease (P(disease) = 0.01), what is the probability that a person has the disease given a positive test result?

Using Bayes' Theorem:
P(disease|positive) = (P(positive|disease) * P(disease)) / P(positive)
P(positive) = P(positive|disease) * P(disease) + P(positive|no disease) * P(no disease)
P(positive) = 0.9 * 0.01 + 0.05 * 0.99 = 0.0595
P(disease|positive) = (0.9 * 0.01) / 0.0595 = 0.1512 or 15.12%

Q5: A confidence interval is a range of values that is likely to contain an unknown population parameter, such as the mean or proportion, with a specified level of confidence.

To calculate a confidence interval for the population mean (μ) with a known population standard deviation (σ) and a sample size (n):

1. Determine the desired level of confidence (e.g., 95%, 99%).
2. Calculate the z-score corresponding to the desired confidence level from the standard normal distribution.
3. Calculate the margin of error as: z * (σ / √n)
4. The confidence interval is calculated as: sample mean ± margin of error

Example:
A random sample of 100 items has a mean of 50 and a standard deviation of 5. Calculate the 95% confidence interval for the population mean.

Given:
- Sample mean = 50
- Sample standard deviation = 5
- Sample size = 100
- Desired confidence level = 95%

Step 1: The z-score for a 95% confidence level is 1.96.
Step 2: Margin of error = 1.96 * (5 / √100) = 0.98
Step 3: The 95% confidence interval is: 50 ± 0.98, or [49.02, 50.98].

This means that we can be 95% confident that the true population mean lies between 49.02 and 50.98.

Q8. The margin of error in a confidence interval is the range of values around the sample statistic that is likely to contain the true population parameter. It is the value added and subtracted from the sample statistic to create the confidence interval.

The formula for the margin of error is:

Margin of Error = z * (standard deviation / √sample size)

Where z is the z-score corresponding to the desired confidence level (e.g., 1.96 for a 95% confidence level).

The sample size affects the margin of error in the following way:

- As the sample size increases, the margin of error decreases.
- Larger sample sizes provide more precise estimates of the population parameter, resulting in a smaller margin of error.

- Sample 2 (n = 100): Average monthly sales = $5,000, standard deviation = $1,000

Q10. To conduct a hypothesis test to determine if the new weight loss drug is significantly effective at a 95% confidence level, we can use a one-sample t-test.

Null hypothesis (H0): The drug is not effective, and the mean weight loss is 0.
Alternative hypothesis (H1): The drug is effective, and the mean weight loss is greater than 0.

Given:
- Sample size (n) = 50
- Sample mean weight loss = 6 pounds
- Sample standard deviation = 2.5 pounds

Step 1: Calculate the test statistic (t-value):
t = (sample mean - hypothesized mean) / (sample standard deviation / √sample size)
t = (6 - 0) / (2.5 / √50) = 16

Step 2: Determine the p-value:
The p-value for a one-tailed t-test with 49 degrees of freedom (n-1) and a t-value of 16 is approximately 0.

Step 3: Compare the p-value to the significance level:
Since the p-value (0) is less than the significance level of 0.05 (95% confidence level), we can reject the null hypothesis.

Conclusion: The results provide strong evidence that the new weight loss drug is significantly effective.
