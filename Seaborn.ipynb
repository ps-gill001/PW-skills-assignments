Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset('fmri')

plt.figure(figsize=(10,8))
plt.title('Line Plot')
sns.lineplot(data = fmri, x = 'timepoint', y= 'signal', hue = 'event', style = 'region')
plt.xlabel('Timepoint')
plt.ylabel('signal')
plt.show()
```

```python
'''Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

Note: pclass, age, and fare are columns in the titanic dataset'''

tit = sns.load_dataset('titanic')
plt.figure(figsize=(8, 6))
plt.subplot(1,2,1)
sns.boxplot(data = tit, x = 'pclass',y='age')
plt.title('Box Plot ')
plt.xlabel('pclass')
plt.ylabel('age')


plt.figure(figsize=(8, 6))
plt.subplot(1,2,2)
sns.boxplot(data = tit, x = 'pclass',y='fare')
plt.title('Box Plot ')
plt.xlabel('pclass')
plt.ylabel('fare')

# Display the plot
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Plot the histogram with hue parameter
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x="price", hue="cut", kde=True)
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
'''Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.'''

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

iris = sns.load_dataset("iris")

sns.pairplot(iris, hue ='species')
plt.title('Pairplot')
plt.show()
```
