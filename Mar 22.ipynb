
### Q1. Pearson Correlation Coefficient for Study Time and Exam Scores

Given data:
- Study time (hours): [3, 5, 1, 6, 4, 2, 7, 5, 3, 4]
- Exam scores: [80, 85, 60, 90, 75, 70, 95, 85, 80, 77]

To calculate Pearson correlation coefficient using Python:

```python
import numpy as np

# Data
study_time = np.array([3, 5, 1, 6, 4, 2, 7, 5, 3, 4])
exam_scores = np.array([80, 85, 60, 90, 75, 70, 95, 85, 80, 77])

# Calculate Pearson correlation coefficient
pearson_corr = np.corrcoef(study_time, exam_scores)[0, 1]

print(f"Pearson correlation coefficient: {pearson_corr:.2f}")
```

**Interpretation**: 
- Pearson correlation coefficient ranges from -1 to 1.
- **Interpretation**: A positive value close to 1 indicates a strong positive linear relationship between study time and exam scores, meaning as study time increases, exam scores tend to increase.

### Q2. Spearman's Rank Correlation for Sleep and Job Satisfaction

Given data:
- Hours of sleep: [7, 6, 8, 5, 7, 6, 9, 5, 8, 6]
- Job satisfaction (scale of 1-10): [8, 7, 9, 5, 8, 6, 9, 4, 7, 6]

To calculate Spearman's rank correlation using Python:

```python
from scipy.stats import spearmanr

# Data
hours_sleep = [7, 6, 8, 5, 7, 6, 9, 5, 8, 6]
job_satisfaction = [8, 7, 9, 5, 8, 6, 9, 4, 7, 6]

# Calculate Spearman's rank correlation coefficient
spearman_corr, _ = spearmanr(hours_sleep, job_satisfaction)

print(f"Spearman's rank correlation coefficient: {spearman_corr:.2f}")
```

**Interpretation**:
- Spearman's rank correlation coefficient also ranges from -1 to 1.
- **Interpretation**: A positive value close to 1 indicates a strong monotonic relationship between hours of sleep and job satisfaction, meaning as hours of sleep increase, job satisfaction tends to increase.

### Q3. Pearson vs Spearman Correlation for Exercise Hours and BMI

Given data:
- Exercise hours per week: [3, 5, 2, 4, 6, 1, 3, 5, 2, 4]
- BMI: [25, 27, 24, 26, 28, 23, 25, 27, 24, 26]

```python
# Data
exercise_hours = np.array([3, 5, 2, 4, 6, 1, 3, 5, 2, 4])
bmi = np.array([25, 27, 24, 26, 28, 23, 25, 27, 24, 26])

# Calculate Pearson correlation coefficient
pearson_corr_exercise_bmi = np.corrcoef(exercise_hours, bmi)[0, 1]

# Calculate Spearman's rank correlation coefficient
spearman_corr_exercise_bmi, _ = spearmanr(exercise_hours, bmi)

print(f"Pearson correlation coefficient (Exercise hours vs BMI): {pearson_corr_exercise_bmi:.2f}")
print(f"Spearman's rank correlation coefficient (Exercise hours vs BMI): {spearman_corr_exercise_bmi:.2f}")
```

**Comparison**:
- **Pearson vs Spearman**: Pearson assumes a linear relationship, while Spearman assesses monotonic relationships (any increasing or decreasing relationship).
- **Interpretation**: If Pearson correlation is significantly different from Spearman's, it suggests the relationship is non-linear.

### Q4. Pearson Correlation Coefficient for TV Watching Hours and Physical Activity

Given data:
- TV watching hours: [3, 4, 2, 5, 3, 4, 6, 2, 4, 5]
- Physical activity level: [2, 3, 1, 4, 2, 3, 5, 1, 3, 4]

**Interpretation**:
- **Interpretation**: A negative correlation close to -1 suggests that as TV watching hours increase, physical activity tends to decrease.

### Q6. Pearson Correlation Coefficient for Sales Calls and Sales Made

Given data:
- Sales calls per day: [10, 8, 12, 9, 11, 7, 13, 9, 10, 8]
- Sales made per week: [2, 3, 1, 4, 2, 3, 5, 1, 3, 4]

```python
# Data
sales_calls = np.array([10, 8, 12, 9, 11, 7, 13, 9, 10, 8])
sales_made = np.array([2, 3, 1, 4, 2, 3, 5, 1, 3, 4])

# Calculate Pearson correlation coefficient
pearson_corr_sales = np.corrcoef(sales_calls, sales_made)[0, 1]

print(f"Pearson correlation coefficient (Sales calls vs Sales made): {pearson_corr_sales:.2f}")
```

**Interpretation**:
- **Interpretation**: A positive correlation close to 1 suggests that as the number of sales calls increases, the number of sales made tends to increase.
